```python
# Function in function

def mean_function(vector):
    item_sum = 0
    for i in vector:
        item_sum += i
        mean = item_sum/len(vector)
    return mean

def variance_function(vector):
    mean = mean_function(vector)
    sumatory = 0
    for i in vector:
        sumatory += (i-mean)**2
        variance = sumatory/len(vector)
    return variance

vector = [1, 2, 3, 4, 5]

print(mean_function(vector))
print(variance_function(vector))


import numpy as np
print(np.mean(vector))
print(np.var(vector))



```
